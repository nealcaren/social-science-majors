```python
major_codes = {
    'Economics'         : [45.06, 45.0601,45.0602,45.0603,45.0604,45.0605,5.0699],
    'Sociology'         : [45.11, 45.1101, 45.12, 45.1201, 45.13, 45.1301, 45.14, 45.1401],
    'Political Science' : [45.1001, 45.1002, 45.1003, 45.1004, 45.1099],
    'Anthropology'      : [45.02, 45.0201, 45.0202, 45.0203, 45.0204, 45.0299, 
                           45.03, 45.0301],
    'Psychology'        : [42.0101, 42.0201, 42.0301, 42.0401, 42.0501, 42.0601, 
                           42.0701, 42.0801, 42.0901, 42.1001, 
                           42.1101, 42.1201, 42.1301, 42.1501, 42.1601, 42.9999],
    'Criminology'       : [43.0101, 43.0102, 43.0103, 43.0104, 43.0105, 43.0106, 
                           43.0107, 43.0108, 43.0109, 43.0199, 43.0201, 43.0202, 
                           43.0203, 43.0299, 43.9999]
}
```

```python
# identify school ids that fit various categories

df = pd.read_csv('data/hd2017.csv', encoding = 'latin1')

# Carnegie codes for BA, MA, and PhD institutions. Eliminates 2-year and speciality schools
school_cats = [15, 16, 21, 22, 31, 32, 33, 60]

schools = list(df[df['CARNEGIE'].isin(school_cats)].UNITID.values)

print(len(schools))
```

Extra the major totals for each year for each major.

```python
def extract(df):
    # Fix columns to create consitency across years
    df.columns = df.columns.str.upper()

    if 'CRACE24 ' in df.keys():
        df['CRACE24'] =  df['CRACE24 ']
    
    if 'CRACE24' not in df.keys() and 'CTOTALT' not in df.keys():
        df['CRACE24'] = df['CRACE15'] + df['CRACE16']
        
    if 'CTOTALT' not in df.keys():
        df['CTOTALT'] = df['CRACE24']
        df['CTOTALM'] = df['CRACE15']
        df['CTOTALW'] = df['CRACE16']


    df = df[df['CIPCODE']  != 99.0000].copy() # Eliminate aggregates
    df = df[df['AWLEVEL']  == 5].copy() # Keep only 4 year undergraduate degrees
    
    df = df[df['UNITID'].isin(schools)] # Eliminate speciality schools 
    # df = df[df['UNITID'].isin(elites)]

    # aggregate stats for total majors and totals for each social science major
    summary = {'Majors'        : df['CTOTALT'].sum(),
               'Majors Male'   : df['CTOTALM'].sum(),
               'Majors Female' : df['CTOTALW'].sum(),
               'Year'          : year}
    
    for major in major_codes:
        summary[major] = df[df['CIPCODE'].isin(major_codes[major])]['CTOTALT'].sum()
        summary['%s Male'   % major] = df[df['CIPCODE'].isin(major_codes[major])]['CTOTALM'].sum()
        summary['%s Female' % major] = df[df['CIPCODE'].isin(major_codes[major])]['CTOTALW'].sum()

    return(summary)
```

```python
totals = []
```

```python
# cycle throug the old files.

for year in [1984, 1985, 1986, 1987, 1988, 1991, 1992, 1993, 1994, 1995, 1996, 1997]:
    print(year)
    csv_name = 'data/%s.txt' % year
    col_names = ['CIPCODE', 'AWLEVEL', 'YEAR','UNITID', 'VARIABLE', 'COUNT']
    df_long = pd.read_csv(csv_name, delimiter='\t', names = col_names )
    
    df = pd.pivot_table(df_long,
                        index=['CIPCODE', 'AWLEVEL', 'YEAR','UNITID' ], 
                        columns = 'VARIABLE',
                        aggfunc=np.sum)

    df = pd.DataFrame(df.to_records())

    df['CRACE15'] = df["('COUNT', 'CRACE15')"]
    df['CRACE16'] = df["('COUNT', 'CRACE16')"]
    df['TOTAL'] = df["('COUNT', 'TOTAL')"]
    
    summary = extract(df)
    totals.append(summary)
```

```python
# Cycle through the new files.

for year in range(1998, 2018):
    print(year)
    csv_name = 'data/c%s_a.csv' % year
    
    df = pd.read_csv(csv_name)
    
    summary = extract(df)
    totals.append(summary)
```

```python
#create data frame
major_df = pd.DataFrame(totals).set_index('Year')
```

Create a new dataframe to be filled with % rather than raw counts.

```python
major_rate_df = major_df.copy()
```

```python
for major in list(major_codes.keys()):
    major_rate_df[major] =  (major_df[major] * 100 )/ major_df['Majors']
```

```python
# Decline as a % from max. 

for field in ['Sociology', 'Political Science']:
    max_rate = major_rate_df[field].max() 
    min_rate = major_rate_df[field].loc[2017]

    decline = 100*(min_rate - max_rate)/max_rate
    print('{0} {1:.0f}% '.format(field, decline))
```

```python
title = "Sociology degrees as share of all degrees"
ax = major_rate_df['Sociology'].plot(title = title, color = '#e5ae38')

ax.set_xlabel('')
ax.set_ylabel('Percent')



print('')
```
